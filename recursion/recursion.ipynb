```python
def countdown(n):

    if n == 0:          # Base case
        print("Done")
        return

    print(n)

    countdown(n - 1)    # Recursive case

countdown(5)
```

```python
def factorial(n):

    if n <= 1:
        return 1

    return n * factorial(n - 1)

print(factorial(5))
```

```python
def total(n):

    if n == 0:
        return 0

    return n + total(n - 1)

print(total(5))
```

```python
def fibonacci(n):

    if n <= 1:
        return n

    return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(6))
```
