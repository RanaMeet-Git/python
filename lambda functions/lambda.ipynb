```python
#normal function
def square(x):
    return x * x


#lambda function
square_lambda = lambda x: x * x

print(square(5))  # Output: 25
print(square_lambda(5)) # Output: 25
```

```python
#lambda with condtional expression
largest = lambda a,b: a if a>b else b

print(largest(50,30))
print(largest(20,30))
```

```python
number = lambda x: "Positive" if x>0 else "Negative"

print(number(-4))
print(number(4))
```

```python
check = lambda x: "Even" if x%2==0 else "Odd"

print(check(7))
print(check(8))
```

```python
#passing lambda to another function
def apply_function(func, value):
    return func(value)

# Example usage
result = apply_function(lambda x: x * 2, 5)
print(result)  
```

```python
students = [
    ("Rana",85),
    ("Rahul",90),
    ("Anu",80)
]

result = sorted(students, key=lambda student: student[1])

print(result)
```

```python
students = [
    ("Rana",85),
    ("Rahul",90),
    ("Anu",80)
]

top = max(students, key=lambda student: student[1])

print(top)


lowest = min(students, key=lambda student: student[1])

print(lowest)
```

```python
#nested lambda function
nested_lambda = lambda x: (lambda y: x + y)
result = nested_lambda(5)(10)  # This will return 15
print(result)
```

```python
multiply = lambda x: lambda y: x*y

result = multiply(5)

print(result(4))
```

```python
#map()
#apply a function to each item in an iterable (like a list) and return a new iterable with the results.
numbers = [1, 2, 3, 4, 5]
squared = map(lambda x: x**2, numbers)
print(list(squared))  # Output: [1, 4, 9, 16, 25]
```

```python
def square(x):
    return x * x

numbers = [1, 2, 3, 4]

result = map(square, numbers)

print(list(result))
```

```python
a = [1,2,3]
b = [10,20,30]

result = map(lambda x,y: x+y, a, b)

print(list(result))
```

```python
#filter
#selects elements from an iterable based on a condition and returns a new iterable with the selected elements.
def even(x):
    return x % 2 == 0

numbers = [1,2,3,4,5,6]

result = filter(even, numbers)

print(list(result))
```

```python
#lambda with filter
numbers = [1,2,3,4,5,6]
result = filter(lambda x: x%2==0, numbers)
print(list(result))
```

```python
marks = [35,80,20,90,45]

passed = filter(lambda m:m>=35, marks)

print(list(passed))
```

```python
#reduce
#applies a rolling computation to sequential pairs of values in an iterable and returns a single value

from functools import reduce

from functools import reduce

numbers = [1,2,3,4]

result = reduce(lambda x,y:x+y, numbers)

print(result)
```

```python
#reduce with lambda
from functools import reduce

numbers = [5,9,2,11,4]

largest = reduce(lambda a,b:a if a>b else b, numbers)

print(largest)
```
