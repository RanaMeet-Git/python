```python
A = {1,2,3,4}
B = {3,4,5,6}

print(A.union(B))
```

```python
#union operator
print(A | B)
```

```python
A = {1,2,3,4}
B = {3,4,5,6}

print(A.intersection(B))
```

```python
#intersection operator
print(A & B)
```

```python
A = {1,2,3,4}
B = {3,4,5,6}

print(A.difference(B))
```

```python
#difference operator
print(A - B)
```

```python
A = {1,2,3,4}
B = {3,4,5,6}

print(A.symmetric_difference(B))
```

```python
#symmetric difference operator
print(A ^ B)
```

```python
#subset 
A = {1, 2, 3}
B = {1, 2, 3, 4, 5}
print(A.issubset(B))
```

```python
#super set
print(B.issuperset(A))
```

```python
#disjoint set
A = {1,2}
B = {3,4}

print(A.isdisjoint(B))
```

```python
#frozenset
#it an immutable set, which means that the elements of a frozenset cannot be changed once it is created.
A = frozenset([1, 2, 3, 4])
print(A)
```
