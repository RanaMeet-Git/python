```python
#without paramerts
def greet():
    print("Hello, Welcome to Python!")

greet()
```

```python
#with parameters
def greet(name):    
    print("Hello, " + name + ". Welcome to Python!")

greet("rana")
```

```python
#with return value
def add_numbers(a, b):
    return a + b
add_numbers(5, 3)
```

```python
#multiple parameters
def greet(name, age):
    print("Hello, " + name + ". You are " + str(age) + " years old.")

greet("rana", 25)
```

```python
#default parameters
def greet(name="guest", age=30):
    print("Hello, " + name + ". You are " + str(age) + " years old.")

greet()
greet("rana", 25)
```

```python
#recusion function
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)


factorial(5)  # Output: 120
```
