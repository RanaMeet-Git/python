```python
#fun inside a variable
def greet():
    return "Hello"

message = greet

print(message())
```

```python
#fun inside a function
def outer_function():
    def inner_function():
        return "Hello from inner function"
    return inner_function
result = outer_function()
print(result())
```

```python
#pure function
#does not have any side effects and always produces the same output for the same input.
def sqare(x):
    return x * x
print(sqare(5))
print(sqare(6))
```

```python
#impure function
#has side effects and may produce different outputs for the same input.
count = 0
def increment():
    global count
    count += 1
    return count

print(increment())
print(increment())
```

```python
name = input("Enter your name: ")

def greeting():
    return "Hello " + name

print(greeting())
```

```python
#higher order function
#A higher-order function is a function that takes one or more functions as arguments, returns a function as its result, or both.
def square(x):
    return x*x

def calculate(function, number):
    return function(number)

print(calculate(square,5))
```

```python
#function compostion 
def double(x):
    return x*2

def square(x):
    return x*x

result = square(double(5))

print(result)
```

```python
#exmple
def bonus(salary):
    return salary+5000

def tax(salary):
    return salary-3000

salary = 50000

salary = bonus(salary)

salary = tax(salary)

print(salary)
```
