```python
import time

def task1():
    print("Preparing coffee...")
    time.sleep(5)
    print("Coffee Ready")

def task2():
    print("Serving Customer")

task1()
task2()
```

```python
import asyncio

async def coffee():

    print("Making Coffee")

    await asyncio.sleep(5)

    print("Coffee Ready")

async def study():

    print("Studying Python")

async def main():

    await asyncio.gather(
        coffee(),
        study()
    )

asyncio.run(main())
```
