```python
#positinal arguments
def greet(name, age):
    print(f"Hello {name}, you are {age} years old.")

greet("rana", 25)
```

```python
#keyword arguments
def greet(name, age):
    print(f"Hello {name}, you are {age} years old.")

greet(name="rana", age=25)
```

```python
#default arguments
def greet(name, age=30):
    print(f"Hello {name}, you are {age} years old.")

greet("rana")  # Output: Hello rana, you are 30 years old.
greet("rana", 25)  # Output: Hello rana, you are 25 years old.
```

```python
#*args and **kwargs
#*args is used to pass a variable number of non-keyword arguments, while **kwargs is used to pass a variable number of keyword arguments.
def numbers(*args):
    print(args)

numbers(10, 20, 30)
```

```python
def total(*numbers):
    print(sum(numbers))

total(10, 20, 30, 40)
```

```python
#kwargs is used to pass a variable number of keyword arguments to a function. It allows you to handle named arguments that you may not have defined in advance.
def student(**details):
    print(details)

student(name="Rana", age=22, city="Hyderabad")
```

```python
def student(**details):
    for key, value in details.items():
        print(key, ":", value)

student(name="Rana", age=22, course="Python")
```

```python
def example(*args, **kwargs):
    print(args)
    print(kwargs)

example(10, 20, 30, name="Rana", city="Hyderabad")
```
