```python
dict={"name": ["John","Doe","Smith"],
      "age": [30, 25, 35],
      "city": ["New York", "Los Angeles", "Chicago"]}
```

```python
for i in range(len(dict["name"])):
    print(f"Name: {dict['name'][i]}, Age: {dict['age'][i]}, City: {dict['city'][i]}")
```

```python
dict.keys() # returns a view object that displays a list of all the keys in the dictionary
```

```python
dict.items() # returns a view object that displays a list of a tuple of all the key-value pairs in the dictionary
```

```python
dict.values() # returns a view object that displays a list of all the values in the dictionary
```

```python
dict.remove("age") # removes the key-value pair with the key "age" from the dictionary
```

```python
print(dict) # prints the updated dictionary without the "age" key-value pair
```

```python
dict.update({"country": ["USA", "USA", "USA"]}) # adds a new key-value pair to the dictionary
print(dict) # prints the updated dictionary with the new "country" key-value pair
```

```python
dict.setdefault("ArithmeticError", "This is a new key-value pair") # adds a new key-value pair to the dictionary if the key does not already exist
print(dict) # prints the updated dictionary with the new "ArithmeticError" key-value pair
```

```python
keys = ["name", "age", "course"]

student = dict.fromkeys(keys, "unknown") # creates a new dictionary with the specified keys and a default value of "unknown"

print(student)
```

```python
#using string 
word = "hello"
word_dict = dict.fromkeys(word, 0) # creates a new dictionary with the characters
print(word_dict) # prints the new dictionary with the characters of the string as keys and 0 as the default value
```

```python
dict.clear() # clears the dictionary
```
