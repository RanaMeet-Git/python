```python
#dunder methods 
#__str__
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def show(self):
        print(self.name,self.age)
    def __str__(self):
        return f"student name:{self.name}"
    def __len__(self):
        return len(self.name)
```

```python
s3=Student("rana",24)
s3.age
```

```python
#__str__ and __repr__
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old."

    def __repr__(self):
        return f"Student(name='{self.name}', age={self.age})"


s = Student("Alice", 20)

print(s)
print(str(s))
print(repr(s))
```

```python
#inheritance
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    pass

d = Dog()
d.speak()
```

```python
#method overridding
class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):
    def speak(self):
        print("Bark")

d = Dog()
d.speak()
```
