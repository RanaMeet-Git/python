```python
#__str__ and __repr__
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old."

    def __repr__(self):
        return f"Student(name='{self.name}', age={self.age})"


s = Student("Alice", 20)

print(s)
print(str(s))
print(repr(s))
```

```python
#inheritance
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    pass

d = Dog()
d.speak()
```

```python
#method overridding
class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):
    def speak(self):
        print("Bark")

d = Dog()
d.speak()
```

```python
#multilevel inheritance
class A:
    def show(self):
        print("A")

class B(A):
    pass

class C(B):
    pass

c = C()
c.show()
```

```python
#multple inheritance
class Father:
    def skill1(self):
        print("Driving")

class Mother:
    def skill2(self):
        print("Cooking")

class Child(Father, Mother):
    pass

c = Child()
c.skill1()
c.skill2()
```

```python
#encapsulation
class Demo:
    def __init__(self):
        self.public = 10
        self._protected = 20
        self.__private = 30

d = Demo()

print(d.public)
print(d._protected)
print(d.__private)   # Error
print(d._Demo__private)  # Possible due to name mangling
```

```python

class Student:

    def __init__(self):

        self.__marks = 0

    def set_marks(self, marks):

        if 0 <= marks <= 100:
            self.__marks = marks
        else:
            print("Invalid Marks")

    def get_marks(self):
        return self.__marks

student = Student()

student.set_marks(95)

print(student.get_marks())
```

```python
student.set_marks(150)
```

```python
#abstraction 
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):

    def start(self):
        print("Car Started")

car = Car()
car.start()
```

```python
#method overloding 
#use *args and *kwargs
class Calculator:

    def add(self, a, b=0, c=0):
        return a + b + c

calc = Calculator()

print(calc.add(5))
print(calc.add(5, 10))
print(calc.add(5, 10, 15))
```

```python
#polymorphism 
class Dog:
    def speak(self):
        print("Bark")

class Cat:
    def speak(self):
        print("Meow")

animals = [Dog(), Cat()]

for animal in animals:
    animal.speak()
```
