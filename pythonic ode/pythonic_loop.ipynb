```python
#enumerate
fruits = ["Apple", "Banana", "Orange"]

for index, fruit in enumerate(fruits):
    print(index, fruit)
    
```

```python
students = ["Rahul", "Amit", "John"]

for roll, student in enumerate(students, start=101):
    print(roll, student)
```

```python
#zip
names = ["John","Alice","Bob"]
marks = [90,85,99]
x=zip(names,marks)
for i in x :
    print(i)
```

```python
names = ["A","B","C"]
marks = [90,80,70]
cities = ["Delhi","Mumbai","Pune"]

for name, mark, city in zip(names, marks, cities):
    print(name, mark, city)
```

```python
names = ["John","Alice","Bob"]
marks = [90,85,99]
x=dict(zip(names,marks))
print(x)
y=list(zip(names,marks))
print(y)
```

```python
#break
for i in range(10):

    if i == 5:
        break

    print(i)
```

```python
#continue
for i in range(5):

    if i == 2:
        continue

    print(i)
```

```python
#else with loop
for i in range(5):
    print(i)

else:
    print("Loop completed")
```

```python
numbers = [10,20,30,40]

target = 25

for num in numbers:

    if num == target:
        print("Found")
        break

else:
    print("Not Found")
```
