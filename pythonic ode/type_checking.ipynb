```python
#int type checking
def add(a: int, b: int) -> int:
    return a + b
```

```python
def greet(name: str) -> str:
    return f"Hello {name}"

print(greet("Alice"))
```

```python
def average(scores: list[int]) -> float:
    return sum(scores) / len(scores)

print(average([80, 90, 100]))
```

```python
def get_age(data: dict[str, int], name: str) -> int:
    return data[name]

students = {
    "John": 21,
    "Alice": 20,
    "Bob": 22
}

print(get_age(students, "Alice"))
```
