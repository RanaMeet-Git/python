```python
#normal method to create list
numbers = []

for i in range(5):
    numbers.append(i)
numbers
```

```python
#comprehension method
numbers = [i for i in range(5)]

numbers
```

```python
#square
sqa=[x*x for x in range (6)]
sqa
```

```python
#even number
even=[x for x in range(20) if x%2==0]
even
```

```python
names = ["john","alice"]

upper = [name.upper() for name in names]

print(upper)
```

```python
matrix = [[1,2],[3,4]]

flat = [item for row in matrix for item in row]

print(flat)
```
