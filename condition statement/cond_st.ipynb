```python
# if conditon 
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even number")
```

```python
#if else condition
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```
