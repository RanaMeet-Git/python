```python
#match 
#only checks the beginning of the string
text2 = "The rain in Spain"
result2 = re.match("The", text2)
print(result2)
```

```python
#full match
#exactly matches the entire string
text3 = "The rain in Spain"
result3 = re.fullmatch("The rain in Spain", text3)

print(result3)
```

```python
#none because the string does not exactly match the pattern whith the whole string
result4 = re.fullmatch(" Spain", text3)
print(result4)
```

```python
import re

text = "I love Python"

print(re.search("Python", text))
print(re.match("Python", text))
print(re.fullmatch("Python", text))
```

```python
#findall returns a list of all matches in the string
import re

text = "Python Java Python C++ Python"

result = re.findall("Python", text)

print(result)
```

```python
#finditer returns an iterator yielding match objects for all matches in the string
import re

text = "Python Java Python"

result = re.finditer("Python", text)

print(result)

for item in result:
    print(item.group())
```

```python
#sub returns a new string with all matches replaced by the specified string
text = "Python is awesome. Python is easy to learn."
result = re.sub("Python", "Java", text)
print(result)
```

```python
#split returns a list of strings split by the occurrences of the pattern
text = "Python, Java, C++, JavaScript"
result = re.split(", ", text)
print(result)
```

```python
#compile returns a pattern object that can be used for matching
text = "Python is awesome. Python is easy to learn."

pattern = re.compile("Python")
result = pattern.sub("Java", text)
print(result)
```

```python
pattern.search(text)
pattern.findall(text)
pattern.match(text)
```

```python
import re

text = "Python Programming"

match = re.search("Python", text)

print(match)
```

```python
print(match.group())
```

```python
#return the start index
print(match.start())
```

```python
#return the end index of the match
print(match.end())
```

```python
#return the start and end positions of the match as a tuple
print(match.span())
```

```python
#flaggs can be used to modify the behavior of the pattern matching
#re.IGNORECASE - ignore case
import re

text = "python"

print(re.search("Python", text, re.IGNORECASE))
```

```python
import re

pattern = re.compile("Python")

reviews = [
    "Python is amazing",
    "I love Python",
    "Python is easy",
    "Java is good"
]

for review in reviews:
    if pattern.search(review):
        print(review)
```
