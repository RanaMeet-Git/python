```python
#appending data to a file
file=open("stdu.txt","a")
file.write("\nfile handling in python is easy")
file.close()
file=open("stdu.txt","r")
print(file.read())
file.close()
```

```python
#exclusive creation, if the file already exists the operation will fail
file=open("stdu.txt","x")
```

```python
file=open("new.txt","a")
file=open("new.txt","w")
file.write("writing data to a file")
file.close()
file=open("new.txt","r")
print(file.read())
```

```python
file=open("stdu.txt","r")
print(file.readline())
print(file.readlines())
file.close()

```

```python
#automatically closing a file after its suite finishes, even if an exception is raised
with open("std.txt","r") as file:
    print(file.read())
```

```python
name=input("Enter name : ")

with open("users.txt","a") as file:
    file.write(name+"\n")

```
