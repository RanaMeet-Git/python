```python


def create_json(filename, dictionary):
    with open(filename, "w") as file:
        json.dump(dictionary, file, indent=4)
```

```python
employee={
"name":["rana","meet","rega"],
"age":[22,25,22]
}
create_json("employees.json",employee)
```

```python
def read_json(filename):
    with open(filename) as file:

        data=json.load(file)
    return data
```
