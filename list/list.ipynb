```python
a=[1, 2, 3]
print(a)
```

```python
#Indexing
for i in range(len(a)):
    print(f"{a[i]} at index {i}")
```

```python
#Slicing
print(a[0:2]) # prints elements from index 0 to 1
print(a[1:3]) # prints elements from index 1 to 2
print(a[::2]) # prints every second element
print(a[::-1]) # prints elements in reverse order
```

```python
a.append(4) # adds 4 to the end of the list
print(a)
```

```python
a.count(2) # counts the number of occurrences of 2 in the list
```

```python
a.insert(1, 5) # inserts 5 at index 1
print(a)
```

```python
a.pop() # removes the last element from the list
print(a)
```

```python
a.remove(2) # removes the first occurrence of 2 from the list
print(a)
```

```python
a.sort() # sorts the list in ascending order
print(a)
```

```python
a.reverse() # reverses the list
print(a)
```

```python
b=a.copy() # creates a shallow copy of the list
print(b)
```

```python
b.append(6) # adds 6 to the end of the copied list
print(b)
print(a) # original list remains unchanged
```
