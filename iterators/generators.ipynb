```python
#normal function
def numbers():
    return [1,2,3]

print(numbers())
```

```python
def numbers():
    yield 1
    yield 2
    yield 3

g=numbers()
print(next(g))
print(next(g))
```

```python
print(next(numbers()))
```

```python
def demo():

    print("First")
    yield 10

    print("Second")
    yield 20

    print("Third")
    yield 30

g = demo()

print(next(g))
print(next(g))
```

```python
#return vs generator
def demo():
    return 10
    return 20

g=demo()
print(g)
print(next(g))
```

```python
def demo():
    yield 10
    yield 20
g = demo()
print(next(g))
print(next(g))
```

```python
#generator expression
numbers = (x*x for x in range(6))
print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
```

```python
#infinite generator
def infinite():

    n = 1

    while True:
        yield n
        n += 1

g = infinite()

print(next(g))
print(next(g))
print(next(g))
```
