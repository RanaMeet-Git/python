```python
#iter function 
#creates an iterator object from an iterable (like a list, tuple, or string).
numbers = [100,200,300]

iterator = iter(numbers)

print(iterator)
print(next(iterator))  # Output: 100
print(next(iterator))  # Output: 200   
```

```python
colors = ["Red","Green","Blue"]

it = iter(colors)

print(next(it))
print(next(it))
print(next(it))
```

```python
#how for loop works internally
numbers = [10,20,30]

iterator = iter(numbers)

while True:
    try:
        num = next(iterator)
        print(num)
    except StopIteration:
        break
```

```python
#stop iteration exception is raised when there are no more items to return from the iterator.
numbers = [1]

it = iter(numbers)

print(next(it))
print(next(it))

```

```python
#handling stop iteration exception
numbers = [1]
it = iter(numbers)
try:
    while True:
        num = next(it)
        print(num)
except StopIteration:
    print("No more items to iterate over.")
```

```python
class Count:

    def __init__(self, limit):
        self.limit = limit
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):

        if self.current > self.limit:
            raise StopIteration

        value = self.current
        self.current += 1

        return value


counter = Count(5)


for i in counter:
    print(i)
```

```python
#with out iterators
with open("data.txt") as f:
    data = f.readlines()
print(data)
```

```python
#wiht iterators
with open("data.txt") as f:
    for line in f:
        print(line)
```

ENUMERATE

```python
#with out enumerate function

fruits = ['apple', 'banana', 'cherry']
index=0
for i in fruits:
    print(index, i)
    index+=1
```

```python
#using enumerate function
for i,fruits in enumerate(fruits):
    print(i, fruits)
```

```python
fruits = ["Apple","Mango","Orange"]


for index, fruit in enumerate(fruits, start=1):
    print(index, fruit)
```

```python
students = ["Rana","Rahul","Anu"]

for roll, student in enumerate(students, start=101):
    print(roll, student)
```
