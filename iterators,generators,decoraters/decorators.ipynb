```python
#basic decorator
def decorator(func):

    def inside():
        print("good morning ")
        print("before calling the function")

        func()

        print("bye bye")
        print("after calling the function")

    return inside

def greet():
    print("Hello World")


greet = decorator(greet)

greet()

```

```python
def decorator(func):

    def wrapper():
        print("Before")
        func()
        print("After")

    return wrapper

@decorator
def greet():
    print("Hello")

greet()
```

```python
def decorator(func):

    def wrapper(*args, **kwargs):
        print("Before")

        result = func(*args, **kwargs)

        print("After")

        return result

    return wrapper


@decorator
def add(a, b, c):
    return a + b + c

print(add(10, 20, 30))

def info(func):

    def wrapper(*args, **kwargs):
        print(f"Function name: {func.__name__}")
        print(f"Arguments: {args}")
        print(f"Keyword Arguments: {kwargs}")

        result = func(*args, **kwargs)

        print(f"Result: {result}")

        return result

    return wrapper

# Apply the info decorator to the add function
add = info(add)
print(add(10, 20, 30))  
```

```python
def decorator1(func):

    def wrapper():
        print("Decorator 1")

        func()

    return wrapper


def decorator2(func):

    def wrapper():
        print("Decorator 2")

        func()

    return wrapper


@decorator1
@decorator2
def greet():
    print("Hello")

greet()
```
