```python
#generator expression
numbers = (x*x for x in range(6))
print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
```

```python
#infinite generator
def infinite():

    n = 1

    while True:
        yield n
        n += 1

g = infinite()

print(next(g))
print(next(g))
print(next(g))
```

```python
#methods of generators
#send() method
def demo():
    value = yield
    print(value)

g = demo()

next(g)          # Start generator

g.send("Hello")
```

```python
def demo():
    value = yield
    print(value)

g = demo()

next(g)

try:
    g.send("Hello")
except StopIteration:
    print("Generator Finished")
```

```python
#close() method
#stopIteration exception is raised when the close() method is called on a generator. This exception can be caught and handled in the generator function to perform any necessary cleanup or finalization tasks before the generator is terminated.

def numbers():
    print("Yielding 1")
    yield 1

    print("Yielding 2")
    yield 2

    print("Yielding 3")
    yield 3

g = numbers()

print(next(g))    # First value

g.close()         # Close generator

print(next(g))    # Error
```

```python
#exxception handling in generators after using close
def demo():
    try:
        while True:
            print("Yielding 1")
            yield 1

            print("Yielding 2")
            yield 2

            print("Yielding 3")
            yield 3
    except GeneratorExit:
        print("Generator Closed!")

g = demo()

print(next(g))
print(next(g))
print(next(g))

g.close()
```

```python
#throw method
#expecting a value to be thrown into the generator, which can be caught and handled within the generator function. This allows for more complex control flow and error handling in generators.
def demo():
    try:
        yield 1
        yield 2
    except ValueError:
        print("ValueError Handled Inside Generator")

g = demo()

print(next(g))

g.throw(ValueError("Error"))
```
