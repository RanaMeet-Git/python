```python
#local variable
def my_function():
    local_var = 10
    print(local_var)

my_function()
print(local_var)  # This will raise an error because local_var is not defined in this scope
```

```python
#Global variable
global_var = 20
def my_function():
    print(global_var)

my_function()
print(global_var)  # This will print 20 because global_var is defined in the global scope
```

```python
#modification of global variable
global_var = 20
def my_function():
    global global_var  # Declare that we are using the global variable
    global_var += 10

my_function()
print(global_var)  # This will print 30 because we modified the global variable inside the
```
